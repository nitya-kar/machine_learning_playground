```python

#impute the complete dataset

def get_missing_data_startegy(df : pd.DataFrame):
  result = []
  for col in df.columns:
    if df[col].isnull().any():
      first_nonexist_data = df[col].isnull().values[0]
      if isinstance(first_nonexist_data, (int, float, str)):
        startegy = 'mean'
      else:
        startegy = 'most_frequent'
      result.append({'column': col, 'strategy': startegy})
  return result
```

```python
get_missing_data_startegy(titanic_df)
#titanic_df.columns
```
