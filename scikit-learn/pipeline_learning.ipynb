```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import  StandardScaler, MinMaxScaler, OrdinalEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import make_column_transformer
from sklearn.metrics import accuracy_score
import pickle
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nitya-kar/machine_learning_playground/refs/heads/main/dataset/ai_job_dataset.csv')
```

```python
df
```

```python
df = df[['experience_level','employment_type', 'years_experience','salary_currency', 'salary_usd']]
```

```python
df
```

```python
df.isnull().sum()
```

```python
encode_tarsform_pipeline_stage = make_column_transformer((OneHotEncoder(), [0,1,3]), remainder='passthrough')
```

```python
scaler_pipeline_stage = make_column_transformer((MinMaxScaler(), [0,1,2,3]), remainder='passthrough')
```

```python
regression_result = RandomForestRegressor()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:,-1], train_size=0.9)
```

```python
y_train
```

```python
pipe = make_pipeline(encode_tarsform_pipeline_stage, scaler_pipeline_stage, regression_result)
```

```python
pipe
```

```python
pipe.fit(X_train, y_train)
```

```python
X_predicton  = pipe.predict(X_test)
```
